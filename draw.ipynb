```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mpi_out = 'mpioutput2.bin'
 
# 读取二进制方式保存的浮点数
data1 = np.fromfile(mpi_out, dtype=np.float64).reshape(150, 150)
 
# 打印读取到的数据
print(data1.shape)
```

```python
plt.figure(figsize=(4,3), dpi=200)
plt.imshow(data1, vmax=100, vmin=0, cmap='bwr')
plt.colorbar(orientation="vertical")
plt.title('parallel')
plt.show()
```

```python
serial_out = 'serialoutput.bin'
 
# 读取二进制方式保存的浮点数
data2 = np.fromfile(serial_out, dtype=np.float64).reshape(150, 150)
 
# 打印读取到的数据
print(data2.shape)
```

```python
plt.figure(figsize=(4,3), dpi=200)
plt.imshow(data2, vmax=100, vmin=0, cmap='bwr')
plt.colorbar(orientation="vertical")
plt.title('serial')
plt.show()
```

```python
# 最大误差
np.max(abs(data1-data2))
```

```python
# 平均误差
np.mean(abs(data1-data2))
```
